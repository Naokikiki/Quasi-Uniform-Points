# Experiment 01: Mesh Ratio Upper Bound

This notebook plots the upper bound of the mesh ratio $\rho(\Lambda)$ for different numbers of points $N$ in dimensions $d=2$ and $d=5$.

The mesh ratio upper bound is given by:
$$\rho(\Lambda) \leq \frac{d\sqrt{d}}{\lambda_1(\Lambda) \cdot \lambda_1(\Lambda^\perp)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import lattice_qmc.utils as utils
import lattice_qmc.korobov as kor 
import lattice_qmc.explicit_rank1 as er1
```

## Compute Mesh Ratio Upper Bounds

## Dimension d=2

```python
# Parameters for d=2
d = 2
N_values_d2 = list(utils.generate_primes(1000, 11))  # generate_primes(n_max, n_min)

print(f"Computing mesh ratio bounds for d={d}...")
print(f"Number of N values: {len(N_values_d2)}")
print(f"N range: {min(N_values_d2)} to {max(N_values_d2)}")
```

```python
# Compute bounds for Korobov lattices (product criterion)
bounds_korobov_d2 = []

for N in N_values_d2:
    lattice = kor.KorobovLattice(d, N, criterion="product")  # KorobovLattice(d, N, ...)
    bounds_korobov_d2.append(lattice.mesh_ratio_upper_bound())

print("Korobov bounds computed.")
```

```python
# Compute bounds for Explicit Rank-1 lattices
# Fix m to get Q_list of length m+1
m = 7

bounds_explicit_d2 = []
Q_values_d2 = []

# For each i, ExplicitRank1Lattice(d, i) gives lattice with Q = Q_list[i]
for i in range(m + 1):
    lattice = er1.ExplicitRank1Lattice(d, i)
    Q = lattice.Q
    if Q < 10:  # Skip small Q
        continue
    bounds_explicit_d2.append(lattice.mesh_ratio_upper_bound())
    Q_values_d2.append(Q)

print("Explicit Rank-1 bounds computed.")
print(f"Q values: {Q_values_d2}")
```

```python
# Plot for d=2
plt.figure(figsize=(10, 6))

plt.plot(N_values_d2, bounds_korobov_d2, 'o-', label='Korobov (product)', markersize=3, alpha=0.7)
plt.plot(Q_values_d2, bounds_explicit_d2, 'd-', label='Explicit Rank-1', markersize=5, alpha=0.7)

plt.xlabel('Number of points N', fontsize=12)
plt.ylabel('Mesh ratio upper bound', fontsize=12)
plt.title('Mesh Ratio Upper Bound (d=2)', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
